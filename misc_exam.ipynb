```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-1, 1)
x, y = np.meshgrid(t, t)
a, b = 0.1, 1
z = a * x**2 + b * y**2 - 1

fig, ax = plt.subplots()
ax.contourf(x, y, z, cmap='viridis')
```
